### Принцип действия

##### Генетический алгоритм — это в первую очередь эволюционный алгоритм, другими словами, основная фишка алгоритма — скрещивание (комбинирование). Как несложно догадаться идея алгоритма наглым образом взята у природы, благо она не подаст на это в суд. Так вот, путем перебора и самое главное отбора получается правильная «комбинация».
##### Алгоритм делится на три этапа:
- Скрещивание
- Селекция (отбор)
- Формирования нового поколения
##### 
##### Если результат нас не устраивает, эти шаги повторяются до тех пор, пока результат нас не начнет удовлетворять или произойдет одно из ниже перечисленных условий:
- Количество поколений (циклов) достигнет заранее выбранного максимума
- Исчерпано время на мутацию
##### 
#### Более подробно о шагах

#### Создание новой популяции. 
##### На этом шаге создается начальная популяция, которая, вполне возможно, окажется не кошерной, однако велика вероятность, что алгоритм эту проблему исправит. Главное, чтобы они соответствовали «формату» и были «приспособлены к размножению».
#### Размножение. 
##### Ну тут все как у людей, для получения потомка требуется два родителя. Главное, чтобы потомок (ребенок) мог унаследовать у родителей их черты. При это размножаются все, а не только выжившие (эта фраза особенно абсурдна, но так как у нас все в сферическом вакууме, то можно все), в противном случае выделится один альфа самец, гены которого перекроют всех остальных, а нам это принципиально не приемлемо.
#### Мутации. 
##### Мутации схожи с размножением, из мутантов выбирают некое количество особей и изменяют их в соответствии с заранее определенными операциями.
#### Отбор. 
##### Тут начинается самое сладкое, мы начинаем выбирать из популяции долю тех, кто «пойдет дальше». При этом долю «выживших» после нашего отбора мы определяем заранее руками, указывая в виде параметра. Как ни печально, остальные особи должны погибнуть.

## Алгоритм роя частиц


##### Чтобы понять алгоритм роя частиц, представьте себе n-мерное пространство (область поиска), в котором рыщут частицы (агенты алгоритма). В начале частицы разбросаны случайным образом по всей области поиска и каждая частица имеет случайный вектор скорости. В каждой точке, где побывала частица, рассчитывается значение целевой функции. При этом каждая частица запоминает, какое (и где) лучшее значение целевой функции она лично нашла, а также каждая частица знает где расположена точка, являющаяся лучшей среди всех точек, которые разведали частицы. На каждой итерации частицы корректируют свою скорость (модуль и направление), чтобы с одной стороны быть поближе к лучшей точке, которую частица нашла сама (авторы алгоритма назвали этот аспект поведения "ностальгией"), и, в то же время, приблизиться к точке, которая в данный момент является глобально лучшей. Через некоторое количество итераций частицы должны собраться вблизи наиболее хорошей точки, хотя возможно, что часть частиц останется где-то в относительно неплохом локальном экстремуме, но главное, чтобы хотя бы одна частица оказалась вблизи глобального экстремума.

## Алгоритм муравья

### Как было сказано выше, муравьиный алгоритм моделирует многоагентную систему. Ее агентов в дальнейшем будем называть муравьями. Как и настоящие муравьи, они довольно просто устроены: для выполнения своих обязанностей они требуют небольшое количество памяти, а на каждом шаге работы выполняют несложные вычисления.

##### Каждый муравей хранит в памяти список пройденных им узлов. Этот список называют списком запретов (tabu list) или просто памятью муравья. Выбирая узел для следующего шага, муравей «помнит» об уже пройденных узлах и не рассматривает их в качестве возможных для перехода. На каждом шаге список запретов пополняется новым узлом, а перед новой итерацией алгоритма – то есть перед тем, как муравей вновь проходит путь – он опустошается.
##### 
##### Кроме списка запретов, при выборе узла для перехода муравей руководствуется «привлекательностью» ребер, которые он может пройти. Она зависит, во-первых, от расстояния между узлами (то есть от веса ребра), а во-вторых, от следов феромонов, оставленных на ребре прошедшими по нему ранее муравьями. Естественно, что в отличие от весов ребер, которые являются константными, следы феромонов обновляются на каждой итерации алгоритма: как и в природе, со временем следы испаряются, а проходящие муравьи, напротив, усиливают их.

## Пчелиный алгоритм


###

#### Алгоритм требует установки ряда параметров, а именно: количество пчел-разведчиков ( ns ), количество сайтов, выбранных из ns посещенных сайтов ( nb ), количество элитных сайтов из nb выбранных сайтов ( ne ), количество пчел, набранных для лучших ne участков ( nre ), количество пчел, набранных для других ( nb-ne ) выбранных участков ( nrb ). Эвристика также требует набора начального размера патчей ( ngh ) и количества циклов, после которых сайт закрывается ( stlim ). Наконец, необходимо определить критерий остановки.

##### Алгоритм начинается со случайного размещения ns пчел-разведчиков в пространстве поиска, а основные этапы алгоритма можно резюмировать следующим образом:

- Оценить приспособленность населения по целевой функции;
- Выберите лучшие сайты для местного (локального) поиска;
- Наберите количество пчел-фуражиров для выбранных участков ( количество пчел для лучших участков ) и оцените их пригодность;
- Выберите наиболее приспособленную пчелу из каждого местного участка в качестве нового центра участка;
- Если сайт не удается улучшить при одном локальном поиске, размер его района уменьшается (сокращение района);
- Если сайт не улучшается в течение интенсивных циклов, сайт закрывается (закрытие сайта);
- Поручите оставшимся пчелам равномерно обыскивать все пространство поиска и оценить их пригодность;
- Если критерий остановки не выполнен, вернитесь к шагу 2;




## Алгоритм дифференциальной эволюции


###
##### Дифференциальная эволюция — это стохастический популяционный метод, который полезен для решения задач глобальной оптимизации. При каждом проходе по совокупности алгоритм изменяет каждое решение-кандидат путем смешивания с другими решениями-кандидатами для создания пробного кандидата. Существует несколько стратегий [3] создания пробных кандидатов, которые подходят для одних задач больше, чем для других. Стратегия best1bin является хорошей отправной точкой для многих систем. В этой стратегии случайным образом выбираются два члена популяции. Их разница используется для мутации лучшего члена («лучшего» в «best1bin»), до сих пор
##### 
##### Затем строится пробный вектор. Начиная со случайно выбранного i-го параметра, испытание последовательно заполняется (по модулю) параметрами исходного кандидата b'или исходного кандидата. Выбор использования b' исходного кандидата или его использование осуществляется с помощью биномиального распределения («bin» в «best1bin») — генерируется случайное число в [0, 1). Если это число меньше константы рекомбинации , то параметр загружается из b', в противном случае — из исходного кандидата. Последний параметр всегда загружается из b'. После того как пробный кандидат создан, оценивается его пригодность. Если испытание лучше первоначального кандидата, то оно занимает свое место. Если он также лучше, чем лучший общий кандидат, он также заменяет его. Чтобы повысить ваши шансы найти глобальный минимум, используйте более высокие значения popsize с более высокими значениями мутации и (сглаживания), но более низкими значениями рекомбинации . Это приводит к расширению радиуса поиска, но замедлению сходимости. По умолчанию лучший вектор решения постоянно обновляется в течение одной итерации ( updating='immediate'). Это модификация [4] исходного алгоритма дифференциальной эволюции, которая может привести к более быстрой сходимости, поскольку пробные векторы могут сразу же получить выгоду от улучшенных решений. Чтобы использовать исходное поведение Storn и Price, обновляя лучшее решение один раз за итерацию, установите updating='deferred'. Этот 'deferred'подход совместим как с распараллеливанием, так и с векторизацией ( 'workers'и 'vectorized'ключевыми словами). Это может повысить скорость минимизации за счет более эффективного использования ресурсов компьютера. Распределение 'workers'вычислений по нескольким процессорам.

```python
import numpy as np
from scipy.optimize import minimize
from geneticalgorithm2 import geneticalgorithm2 as ga
import time
times = []
# Определяем функцию Розенброка
def rosenbrock_function(x):
    return sum(100*(x[1:]-x[:-1]**2)**2 + (1-x[:-1])**2)

# Функция для оценки времени работы эволюционного алгоритма
def evaluate_genetic_algorithm_time(runs):
    times = []
    for _ in range(runs):
        start_time = time.time()
        model.run()
        end_time = time.time()
        times.append(end_time - start_time)
    mean_time = np.mean(times)
    return mean_time

# Выполняем 100 прогонов для каждого из 2 алгоритмов
runs = 100
results_powell = []
results_genetic = []
for _ in range(runs):
    # Минимизируем функцию с помощью метода Пауэлла
    result_powell = minimize(rosenbrock_function, [0, 0], method='powell')
    results_powell.append(result_powell.fun)
    
    # Определяем параметры для генетического алгоритма
    varbound = np.array([[-5, 5]]*2)
    model = ga(function=rosenbrock_function, dimension=2, variable_type='real', variable_boundaries=varbound)
    
    # Запускаем генетический алгоритм
    start_time = time.time()
    result_genetic = model.run(no_plot=True, disable_printing=True)
    end_time = time.time()
    times.append(end_time - start_time)
    results_genetic.append(model.output_dict['function'])

# Вычисляем математическое ожидание и дисперсию для финального значения функции соответствия
mean_powell = np.mean(results_powell)
variance_powell = np.var(results_powell)
mean_genetic = np.mean(results_genetic)
variance_genetic = np.var(results_genetic)

# Выводим результаты
```

```python
print("Method: Powell")
print("Mean:", mean_powell)
print("Variance:", variance_powell)
print("")

print("Genetic Algorithm")
print("Mean:", mean_genetic)
print("Variance:", variance_genetic)

# Оцениваем время работы эволюционного алгоритма
mean_time = np.mean(times)
print("Mean time for Genetic Algorithm:", mean_time, "seconds")
```

```python

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from deap import base, creator, tools, algorithms
import random

# Загрузка данных
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Определение функции оценки (fitness function)
def evaluate_SVM(params):

    C, gamma = params
    if C <0: C=C*-1
    if gamma <0:gamma=gamma*-1
    clf = SVC(C=C, gamma=gamma)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    return accuracy_score(y_test, y_pred),

# Создание объекта для оптимизации
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

# Инициализация параметров генетического алгоритма
toolbox = base.Toolbox()
toolbox.register("attr_float", random.uniform, 0.1, 10)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=2)
# Теперь вызов toolbox.individual()будет вызываться initRepeat()с фиксированными аргументами и возвращать полный creator.
# Individual набор IND_SIZEчисел с плавающей запятой с максимизирующим единственным целевым fitnessатрибутом.
toolbox.register("population", tools.initRepeat, list, toolbox.individual) # создаем популяцию 
toolbox.register("evaluate", evaluate_SVM) # фунгнкцитя SVM 
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)

# Оптимизация параметров с помощью генетического алгоритма
population = toolbox.population(n=30)
algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=10, verbose=False)

# Получение лучших параметров
best_params = tools.selBest(population, k=1)[0]
best_C, best_gamma = best_params
print("Лучшие параметры: C =", best_C, ", gamma =", best_gamma)
```
